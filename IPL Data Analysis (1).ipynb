# Most Wins In IPL

```python
temp=pd.DataFrame({"winner":df["winner"]})

count_win=temp["winner"].value_counts()

labels=[x for x in count_win.keys()]

fig=plt.figure(figsize=(20,15))

plot=plt.pie(x=count_win,labels=labels,autopct="%.1f%%")

plt.title("Most Wins in IPL")
```

# Most Wins In Eliminator

```python
# countplot() method is used to Show the counts of observations in each categorical bin using bars.

sns.countplot(df["winner"][df["eliminator"]=="Y"],data=df)

plt.title("Most Wins In Eliminator",fontsize=20)

plt.xticks(rotation=90)  # rotae the x axis label to 90 degree

plt.show()
```

# Toss Decision

```python
team=df["toss_winner"].unique()

decision=pd.DataFrame([],columns=["Toss Winner","Decision","Times"])

for id,element in enumerate(team):
    
    temp_bat=df[(df["toss_winner"]==element) & (df["toss_decision"]=="bat")]
    
    temp_field=df[(df["toss_winner"]==element) & (df["toss_decision"]=="field")]
    
    # append these elements in list
    
    decision=decision.append({"Toss Winner":element,"Decision":"bat",
                              "Times":temp_bat["toss_winner"].count()},ignore_index=True)
    
    decision=decision.append({"Toss Winner":element,"Decision":"field",
                              "Times":temp_field["toss_winner"].count()},ignore_index=True)
    
```

```python
decision
```

```python
sns.catplot(x="Toss Winner",y="Times",hue="Decision",data=decision,kind="bar",height=5,aspect=2)  # aspect will be the spaces

plt.xticks(rotation=90)

plt.title("Toss Decision Of Teams")

plt.xlabel("IPL Teams")

plt.ylabel("Toss Decision")

plt.show()
```
